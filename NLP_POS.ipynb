```
import pandas as pd
import numpy as np
import nltk
import re
import string
import zipfile as zipf
from nltk.corpus import stopwords
from nltk import WordNetLemmatizer
from sklearn.feature_extraction.text import CountVectorizer, TfidfVectorizer
from string import punctuation
from google.colab import drive
drive.mount('/content/drive')
```

```
database_path = "/content/drive/My Drive/DBs/shakespeare/"
extract_folder = zipf.ZipFile(database_path+'shakespeare_data.zip')
book = extract_folder.read('shakespeare_data/1kinghenryiv.txt').strip().decode("utf-8")
```

LOWERING and REMOVING PUNCTUATION and NUMERIC data

```
book = re.sub("\n"," ", book).lower()
print(book)
tfidf_vect = TfidfVectorizer()
tfidf_vect.fit_transform([book])
print(tfidf_vect.vocabulary_)
#book = re.sub(r'[^\w\s]', '', book) # only text and spaces
book = re.sub(r'[^a-z\s]', '', book)
print(book)
```

TOKENIZING using NLTK WORD_TOKENIZE

```
nltk.download('punkt')
nltk_tokens = nltk.word_tokenize(book, language='english')
print(nltk_tokens)
```

Removing STOPWORDS

```
nltk.download('stopwords')
stop_words = stopwords.words('english')
final_token = [word for word in nltk_tokens if word not in stop_words]
print(final_token)
print(len(final_token))
```

```
nltk.download('wordnet')
lemmatizer = WordNetLemmatizer()
final_token = [lemmatizer.lemmatize(word) for word in final_token]
print(lemmatizer.lemmatize('frighted'))
```
